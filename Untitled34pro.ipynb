```python
IMG_SIZE = 224
BATCH_SIZE = 32

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    zoom_range=0.15,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True,
    validation_split=0.2
)

train_gen = train_datagen.flow_from_directory(
    IMAGE_DATA_PATH,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    subset='training'
)

val_gen = train_datagen.flow_from_directory(
    IMAGE_DATA_PATH,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    subset='validation'
)
```

```python
def build_explainable_model():
    # Base model: EfficientNetB0
    base_model = EfficientNetB0(weights='imagenet', include_top=False, input_shape=(IMG_SIZE, IMG_SIZE, 3))
    base_model.trainable = False # Freeze base layers

    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(256, activation='relu')(x)
    x = Dropout(0.5)(x)
    predictions = Dense(1, activation='sigmoid')(x)

    model = Model(inputs=base_model.input, outputs=predictions)
    model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
    return model, base_model

model, base_model = build_explainable_model()
print("Model Summary:")
model.summary()
```

```python
EPOCHS = 10
history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=EPOCHS
)

# Save the model
model.save("deepfake_detector_model.h5")
```

```python
def plot_results(history):
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    epochs_range = range(len(acc))

    plt.figure(figsize=(14, 5))
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, acc, label='Training Accuracy', linewidth=2)
    plt.plot(epochs_range, val_acc, label='Validation Accuracy', linewidth=2)
    plt.title('Training and Validation Accuracy')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, loss, label='Training Loss', linewidth=2)
    plt.plot(epochs_range, val_loss, label='Validation Loss', linewidth=2)
    plt.title('Training and Validation Loss')
    plt.legend()
    plt.show()

plot_results(history)
```

```python
def get_gradcam(img_array, model, last_conv_layer_name):
    grad_model = tf.keras.models.Model(
        [model.inputs], [model.get_layer(last_conv_layer_name).output, model.output]
    )
    with tf.GradientTape() as tape:
        last_conv_layer_output, preds = grad_model(img_array)
        class_channel = preds[:, 0]

    grads = tape.gradient(class_channel, last_conv_layer_output)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    last_conv_layer_output = last_conv_layer_output[0]
    heatmap = last_conv_layer_output @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    heatmap = tf.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)
    return heatmap.numpy()

# Test on one image
sample_img_path = next(val_gen)[0][0] # Get a random sample from validation
img_input = np.expand_dims(sample_img_path, axis=0)
heatmap = get_gradcam(img_input, model, "top_activation")

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.imshow(sample_img_path)
plt.title("Original Image")
plt.subplot(1, 2, 2)
plt.imshow(sample_img_path)
plt.imshow(cv2.resize(heatmap, (224, 224)), alpha=0.6, cmap='jet')
plt.title("Explainable Heatmap (Grad-CAM)")
plt.show()
```

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

def predict_video(video_path, model, img_size=224, frames_to_sample=10):
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print(f"Error: Could not open video {video_path}")
        return "Error", 0.0, []

    predictions = []
    sampled_frames = []
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT)) # Corrected attribute here

    if frame_count == 0:
        print(f"Error: Video {video_path} has no frames.")
        return "Error", 0.0, []

    # Determine which frames to sample
    frame_indices = np.linspace(0, frame_count - 1, frames_to_sample, dtype=int)

    for i in range(frame_count):
        ret, frame = cap.read()
        if not ret:
            break

        if i in frame_indices:
            # Preprocess the frame for the model
            img = cv2.resize(frame, (img_size, img_size))
            img = img / 255.0  # Normalize to [0, 1]
            img_input = np.expand_dims(img, axis=0) # Add batch dimension

            # Make prediction
            prediction = model.predict(img_input, verbose=0)[0][0]
            predictions.append(prediction)
            sampled_frames.append(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))

    cap.release()

    if not predictions:
        print("No frames were processed.")
        return "Error", 0.0, []

    # Aggregate predictions (e.g., by averaging or majority vote)
    avg_prediction = np.mean(predictions)
    label = "Fake" if avg_prediction > 0.5 else "Real"

    return label, avg_prediction, sampled_frames


# 1. Set the path to your video file
video_file = "/content/drive/MyDrive/deepfake project/archive (3)/Celeb-synthesis/id0_id16_0003.mp4"

# 2. Check if the file exists before running
if not os.path.exists(video_file):
    print(f"Error: Video file not found at {video_file}")
else:
    # 3. Call the prediction function
    label, score, frames = predict_video(video_file, model)

    # 4. Print the final result
    print("-" * 30)
    print(f"VIDEO ANALYSIS COMPLETE")
    print(f"Final Prediction: {label}")
    print(f"Confidence Score: {score:.4f}")
    print("(Score > 0.5 means Fake, < 0.5 means Real)")
    print("-" * 30)

    # 5. Visualize some frames from the video
    plt.figure(figsize=(20, 10))
    # Display up to 5 frames to see the detail
    num_display = min(5, len(frames))
    for i in range(num_display):
        plt.subplot(1, num_display, i + 1)
        plt.imshow(frames[i])
        plt.title(f"Frame {i+1}\nLabel: {label}")
        plt.axis('off')
    plt.show()
```

### Mathematical Equations and Concepts in this Project

1.  **Image Normalization**
    Pixel values (typically 0-255) are scaled to a range of 0-1:
    $$P_{\text{normalized}} = \frac{P_{\text{original}}}{255}$$

2.  **Binary Cross-entropy Loss**
    This loss function is used for binary classification tasks. For a single sample with true label $y$ (0 or 1) and predicted probability $p$ (between 0 and 1):
    $$L(y, p) = - (y \cdot \log(p) + (1 - y) \cdot \log(1 - p))$$
    The total loss for a batch is typically the average of this loss over all samples.

3.  **Adam Optimizer**
    Adam (Adaptive Moment Estimation) is an optimization algorithm that adjusts the learning rate for each parameter based on estimates of the first and second moments of the gradients. While its full equations are complex, the core idea involves dynamically adapting the learning rate for different parameters during training, generally described as:
    *   Estimating the first moment (mean) of the gradients.
    *   Estimating the second moment (uncentered variance) of the gradients.
    *   Using these estimates to update the model weights.

4.  **Grad-CAM (Gradient-weighted Class Activation Mapping)**
    Grad-CAM is used to generate explainable heatmaps by localizing important regions in the image for a specific prediction. It involves several steps:
    *   **Calculate Gradients**: Gradients of the target class score ($Y^c$) with respect to the feature maps ($A^k$) of a convolutional layer:
        $$\frac{\partial Y^c}{\partial A^k_{ij}}$$
        where $A^k_{ij}$ is the activation of feature map $k$ at spatial location $(i,j)$.
    *   **Global Average Pooling (GAP) of Gradients**: Compute the neuron importance weight ($\alpha_k^c$) for each feature map $k$ by averaging the gradients over its spatial dimensions:
        $$\alpha_k^c = \frac{1}{Z} \sum_i \sum_j \frac{\partial Y^c}{\partial A^k_{ij}}$$
        where $Z$ is the total number of pixels in the feature map.
    *   **Weighted Combination and ReLU**: Combine the weights with the feature maps and apply a ReLU (Rectified Linear Unit) activation to get the final heatmap ($L_{\text{Grad-CAM}}^c$):
        $$L_{\text{Grad-CAM}}^c = \text{ReLU}\left(\sum_k \alpha_k^c A^k\right)$$

5.  **Video Prediction Aggregation**
    When predicting whether a video is fake or real, individual frame predictions ($P_n$) are averaged to produce an overall video prediction score:
    $$\text{Average Prediction} = \frac{1}{N} \sum_{n=1}^{N} P_n$$
    where $N$ is the number of sampled frames and $P_n$ is the model's prediction for frame $n$.
